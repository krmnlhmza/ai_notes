```python
import numpy as np
import matplotlib.pyplot as plt
"""
'o' Circle

'*' Star

'.' Point

',' Pixel

'x' X

'X' X (filled)

'+' Plus

'P' Plus (filled)

's' Square

'D' Diamond

'd' Diamond (thin)

'p' Pentagon

'H' Hexagon

'h' Hexagon

'v' Triangle Down

'^' Triangle Up

'<' Triangle Left

'>' Triangle Right

'1' Tri Down

'2' Tri Up

'3' Tri Left

'4G' Tri Right

'|' VLine

'_' HLine
"""


x = np.array([1,6])
y = np.array([1,60])
plt.plot(x,y,marker="o")  #3. parametre "o" verirsen nokta işaretler
plt.plot(x, y, marker="*", linestyle="-.", color="g")
plt.plot(x,y,"*:r")  #daha kısa hali
plt.plot(x,y,"*:r",mfc="k", linewidth='2.5',mec="b",markersize=30) #direk ms=20 de yazabilrisin
#not marketfacecolor, mfc markırın iç rengini belirler
#not markeredgecolor, mec markerın kenar renginibelirler
"""
Line Syntax	Description
linestyle='solid' (veya ls='-')

linestyle='dotted' (veya ls=':')

linestyle='dashed' (veya ls='--')

linestyle='dashdot' (veya ls='-.')

linestyle='None' (veya ls=' ')
"""
"""
linewidth veya lw

linewidth='20.5'       genişlik

"""
"""
Color Syntax	Description
'r'	Red
'g'	Green
'b'	Blue
'c'	Cyan
'm'	Magenta
'y'	Yellow
'k'	Black
'w'	White
"""
```

```python
#plt.xlabel(), plt.ylabel() ve plt.title()

# Başlık için ayrı, etiketler için ayrı font sözlükleri tanımlanıyor
font_baslik = {'family': 'serif', 'color': 'blue', 'size': 20}
font_etiket = {'family': 'serif', 'color': 'darkred', 'size': 15}
x = np.array([80, 85, 90, 95])
y = np.array([240, 250, 260, 270])

plt.plot(x, y)

plt.title("Spor Veri Takibi", fontdict=font_baslik, loc='center') # Ana başlık left ile de sola aldık, center ile orta vb.
plt.xlabel("Ortalama Nabız", fontdict=font_etiket) # X ekseni etiketi
plt.ylabel("Kalori Yakma", fontdict=font_etiket)  # Y ekseni etiketi
plt.show()
```

```python
#Grafiğe Izgara (Grid) Ekleme (plt.grid())

plt.plot(x, y)
plt.title("Spor Veri Takibi")
plt.xlabel("Ortalama Nabız")
plt.ylabel("Kalori Yakma")

plt.grid() # <<< YENİ ÖĞRENİLEN KOMUT
#plt.grid(axis='y') Sadece Y eksenine (kalori) ait yatay çizgileri göster

# Ana veri çizgilerinin (plot) önüne geçmemesi için
# soluk, ince ve kesikli bir ızgara stili:
plt.grid(color='gray', linestyle='--', linewidth=0.5)
plt.show()
```

```python
#1. Alt Grafikler (plt.subplot())

plt.suptitle("MY SHOP") # Tüm pencerenin ana başlığı

# --- Grafik 1 (Sol) ---
x1, y1 = [0,1,2,3], [3, 8, 1, 10]
plt.subplot(1, 2, 1) # "1 satır, 2 sütunluk alanın 1. hücresini (sol taraf) seç."
plt.plot(x1, y1)
plt.title("Grafik 1")

# --- Grafik 2 (Sağ) ---
x2, y2 = [0,1,2,3], [10, 20, 30, 40]
plt.subplot(1, 2, 2) # "1 satır, 2 sütunluk alanın 1. hücresini (sağ taraf) seç."
plt.plot(x2, y2)
plt.title("Grafik 2")
plt.axis('off')#etraftaki çerçevee gider
plt.show()
```

```python
import matplotlib.pyplot as plt
import matplotlib.patches as patches


fig, ax = plt.subplots()

rect = patches.Rectangle((0.1, 0.2), 5.0, 0.5,  facecolor='lightblue', edgecolor='black')
ax.add_patch(rect)

ax.text(0.15, 0.3, "selam", fontsize=15, color='darkred')

plt.xlim(0, 1)
plt.ylim(0, 1)

plt.show()
```

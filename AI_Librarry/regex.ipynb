```python
import re
```

```python
cumle = "Mustafa Kemal Atatürk, Türk asker, devlet adamı ve Türkiye Cumhuriyeti'nin kurucusudur."

patern = "Türk"
         
re.search(patern, cumle)
```

```python
durum = re.search(patern, cumle)
durum.span()
```

```python
durum.start()
durum.end()
durum.group()
```

```python
for i in re.findall(patern,cumle):
    print(i)
```

```python
for i in re.finditer(patern,cumle): #findall da o kelimeye bakıyor finditerde ise alayına bakar.
    print(i.span(),i.group())
```

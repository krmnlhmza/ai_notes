```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # 'pls' yerine 'plt' standarttır
import matplotlib

# 1. Veriyi oku
bilgi = pd.read_csv("C:\\Users\\user\\Downloads\\cars.csv")

# 2. Grafik için veriyi hazırla: "sales" sütunundaki değerleri say
# Bu, size hangi satış değerinden kaçar tane olduğunu (veriler)
# ve bu değerlerin ne olduğunu (etiketler) verir.
sales_counts = bilgi["sales"].value_counts()

# 3. Grafiği çiz
plt.figure(figsize=(8,8))
plt.style.use("fivethirtyeight")

# DÜZELTME: Grafiğe tüm 'bilgi' tablosunu değil,
# sadece 'sales_counts' sonuçlarını ver.
plt.pie(
    sales_counts.values,  # Dilimlerin boyutları (sayılar, örn: [450, 300, 150])
    labels=sales_counts.index,  # Dilimlerin etiketleri (örn: ["low", "med", "high"])
    autopct="%1.1f%%"     # Yüzdeleri göster
)

plt.title("Satış Değerlerinin Dağılımı") # Başlık eklemek iyi bir fikirdir
plt.show()
```

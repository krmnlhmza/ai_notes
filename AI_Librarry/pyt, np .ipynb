```python
ary= np.array([[[1,2,3,],[132,222,35]],[[1,2,3,],[1432,142,344]]])
print(ary)
print("-"*18)
print(np.flip(ary,axis=0))#axis=1 dersen simetriyi farklı yerden alır
print("-"*18)
print(np.flipud(ary)),
print("-"*18)
print(ary.T) #satırlar sütun olur
```

```python
ary2= np.array([[[1,2,3,],[132,222,35]],[[1,2,3,],[1432,142,344]]])
for a in np.nditer(ary2):
    print(a)
```

```python
ary2= np.array([[[1,2,3,],[132,222,35]],[[1,2,3,],[1432,142,344]]])  #türünü değiştirip alcaz
for a in np.nditer(ary2,flags=['buffered'],op_dtypes=['S']):
    print(a)
```

```python
ary2= np.array([[[0,2,3,],[132,222,35]],[[1,0,3,],[0,142,344]]])  #türünü değiştirip alcaz
for a in np.nditer(ary2,flags=['buffered'],op_dtypes=[np.bool_],casting='unsafe'): #veri kaybolur diye uyarı almayalım diye unsafe yaptık.
    print(a)
```
